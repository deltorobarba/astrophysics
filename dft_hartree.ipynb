Perform Density Functional Theory (DFT) calculations using Python. One of the most popular tools to do so is PySCF, which is a Python module for quantum chemistry calculations, including DFT.

Example of how you can perform a DFT calculation for a water molecule (H₂O) using PySCF:

```python
from pyscf import gto, dft

# Define the molecule
mol = gto.M(
    atom = '''
        O  0.0000000   0.0000000   0.0000000
        H  0.7586022   0.0000000   0.5042847
        H -0.7586022   0.0000000   0.5042847
    ''',
    basis = 'cc-pvdz',  # Use a common basis set
)

# Perform a DFT calculation using the B3LYP functional
mf = dft.RKS(mol)  # Restricted Kohn-Sham DFT for a closed-shell system
mf.xc = 'b3lyp'    # Exchange-correlation functional
energy = mf.kernel()

print(f"DFT energy for H2O using B3LYP: {energy} Hartree")
```

This script outputs the DFT energy in Hartree units.

1. Molecule Definition: The molecule is defined using the gto.M() function where you can provide atomic positions in angstroms and the basis set (in this case, cc-pvdz).
2. DFT Calculation: dft.RKS is used to set up a restricted Kohn-Sham DFT calculation, appropriate for closed-shell systems (like water).
3. Functional: The exchange-correlation functional is set using mf.xc = 'b3lyp', which is one of the popular hybrid functionals.
4. Energy Calculation: The mf.kernel() function performs the actual DFT calculation and returns the energy.

You can change the molecule, basis set, or DFT functional as needed for more complex systems or calculations.


**Achtung:** the output in the code example provided is not the Hamiltonian. Instead, it is the total energy of the system calculated using the DFT method. The Hamiltonian itself is the operator that governs the system, but the mf.kernel() method in PySCF computes the final ground-state energy after solving the Kohn-Sham equations for the system.
