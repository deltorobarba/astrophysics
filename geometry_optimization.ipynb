# **Geometry Optimization in Chemistry**

In PySCF, geometry optimization can be done using an interface to external optimizers like SciPy or by using PySCF's built-in geometric_solver function. For this example, I'll show how to perform a geometry optimization on a simple molecule like ammonia (NH₃) using PySCF's built-in tools.

Geometry optimization is useful because the initial guess for the atomic positions of a molecule might not correspond to the lowest energy configuration. Through optimization, you refine these positions to get the most stable configuration, which is typically the geometry found in experiments.

```python
from pyscf import gto, dft
from pyscf.geomopt import geometric_solver

# Define the molecule with an initial guess for the geometry
mol = gto.M(
    atom = '''
        N  0.0000000   0.0000000   0.2000000
        H  0.0000000   0.9433030  -0.2000000
        H  0.8164970  -0.4716510  -0.2000000
        H -0.8164970  -0.4716510  -0.2000000
    ''',
    basis = 'cc-pvdz',  # Use a standard basis set
    verbose = 4,        # Verbosity level for detailed output
)

# Perform a DFT calculation using the B3LYP functional
mf = dft.RKS(mol)
mf.xc = 'b3lyp'  # Define the exchange-correlation functional

# Perform geometry optimization
mol_eq = geometric_solver(mf)

# Print the optimized geometry
print("Optimized Geometry (in Angstroms):")
for atom, coord in zip(mol_eq.atom, mol_eq.atom_coords()):
    print(f"{atom}: {coord}")
```

Steps Breakdown:
1. Molecule Definition: The geometry of the molecule (ammonia, NH₃) is defined with initial atomic positions. These positions don't need to be exact, as the optimization will refine them.
2. DFT Calculation: We use the B3LYP functional for the DFT calculation. You can choose other functionals based on your problem.
3. Geometry Optimization: The geometric_solver(mf) function is used to perform the geometry optimization. This function iteratively adjusts the positions of the atoms to minimize the total energy of the system.
4. Optimized Geometry: After the optimization is complete, the final geometry (in Angstroms) is printed. This geometry corresponds to the lowest energy configuration found during the optimization process.

```python
from pyscf import gto, dft
from pyscf.geomopt.geometric_solver import optimize  # Correct import

# Define the molecule with an initial guess for the geometry
mol = gto.Mole()
mol.atom = '''
    N  0.0000000   0.0000000   0.2000000
    H  0.0000000   0.9433030  -0.2000000
    H  0.8164970  -0.4716510  -0.2000000
    H -0.8164970  -0.4716510  -0.2000000
'''
mol.basis = 'cc-pvdz'
mol.verbose = 4
mol.build()

# Perform a DFT calculation using the B3LYP functional
mf = dft.RKS(mol)
mf.xc = 'b3lyp'  # Define the exchange-correlation functional

# Perform geometry optimization
mol_eq = optimize(mf)

# Print the optimized geometry
print("Optimized Geometry (in Angstroms):")
for atom, coord in zip(mol_eq.atom, mol_eq.atom_coords()):
    print(f"{atom}: {coord}")
```

**Using SciPy for Optimization (Alternative Method)**

If you want to use SciPy for optimization, here’s how you can do it using the optimize.minimize method from SciPy:

```python
from pyscf import gto, dft
from pyscf.geomopt import minimize
import numpy as np

# Define the molecule
mol = gto.M(
    atom = '''
        N  0.0000000   0.0000000   0.2000000
        H  0.0000000   0.9433030  -0.2000000
        H  0.8164970  -0.4716510  -0.2000000
        H -0.8164970  -0.4716510  -0.2000000
    ''',
    basis = 'cc-pvdz',  # Use a standard basis set
    verbose = 4,        # Verbosity level for detailed output
)

# DFT calculation with B3LYP functional
mf = dft.RKS(mol)
mf.xc = 'b3lyp'

# Perform geometry optimization using SciPy
mol_eq = minimize(mf)

# Print optimized coordinates
print("Optimized Geometry (in Angstroms):")
print(mol_eq.atom_coords())
```

```python
from pyscf import gto, dft
from scipy.optimize import minimize
import numpy as np

# Define the molecule
mol = gto.Mole()
mol.atom = '''
    N  0.0000000   0.0000000   0.2000000
    H  0.0000000   0.9433030  -0.2000000
    H  0.8164970  -0.4716510  -0.2000000
    H -0.8164970  -0.4716510  -0.2000000
'''
mol.basis = 'cc-pvdz'  # Use a standard basis set
mol.verbose = 4        # Verbosity level for detailed output
mol.build()

# Perform a DFT calculation using the B3LYP functional
mf = dft.RKS(mol)
mf.xc = 'b3lyp'  # Define the exchange-correlation functional
mf.kernel()

# Function to calculate total energy from coordinates (to be optimized)
def calc_energy(coords):
    mol.set_geom_(coords, unit='Angstrom')
    mf = dft.RKS(mol)
    mf.xc = 'b3lyp'
    return mf.kernel()

# Initial guess for the atomic positions in a flat array
coords = mol.atom_coords()

# Flatten the coordinates for use in the optimizer
initial_coords = coords.flatten()

# Perform geometry optimization using SciPy's minimize function
result = minimize(calc_energy, initial_coords, method='BFGS')

# Reshape the optimized coordinates back into atomic positions
optimized_coords = result.x.reshape(-1, 3)

# Set the optimized geometry in the molecule object
mol.set_geom_(optimized_coords, unit='Angstrom')

# Print the optimized geometry
print("Optimized Geometry (in Angstroms):")
for atom, coord in zip(mol.atom, mol.atom_coords()):
    print(f"{atom}: {coord}")
```

* minimize(mf): This function wraps the optimization process using SciPy's optimize.minimize to find the optimal geometry.
