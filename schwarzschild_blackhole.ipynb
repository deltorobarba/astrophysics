# **Schwarzschild Black Hole**

This code provides a simplified model of a non-rotating black hole (Schwarzschild black hole). Real black holes often rotate, which leads to more complex physics.

The code focuses on the kinematics (velocities) near a black hole. More advanced research would involve studying the effects of gravity on light (gravitational lensing) and the behavior of matter near the event horizon.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
G = 6.674e-11  # Gravitational constant in N m^2/kg^2
c = 299792458  # Speed of light in m/s
M = 1.989e30  # Mass of the black hole (in solar masses)
M_bh = M * 1.989e30  # Convert to kg

# Schwarzschild radius (event horizon)
r_s = 2 * G * M_bh / c**2

# Define a range of radial distances
r = np.linspace(1.1 * r_s, 10 * r_s, 1000)

# Calculate the orbital velocity for a circular orbit
v_orb = np.sqrt(G * M_bh / r)

# Calculate the escape velocity
v_esc = np.sqrt(2 * G * M_bh / r)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(r / r_s, v_orb / c, label='Orbital Velocity')
plt.plot(r / r_s, v_esc / c, label='Escape Velocity')
plt.axhline(y=1, color='r', linestyle='--', label='Speed of Light')
plt.xlabel('Radial Distance (in Schwarzschild radii)')
plt.ylabel('Velocity (in units of c)')
plt.title('Orbital and Escape Velocities Near a Black Hole')
plt.legend()
plt.grid(True)
plt.show()
```

1. Import Libraries:
    * numpy for numerical calculations.
    * matplotlib.pyplot for plotting.
2. Define Constants:
    * G: Gravitational constant.
    * c: Speed of light.
    * M: Mass of the black hole (assumed to be one solar mass in this example).
    * M_bh: Converts the mass to kilograms.
3. Schwarzschild Radius:
    * Calculates the Schwarzschild radius (r_s), which represents the event horizon of the black hole.
4. Radial Distances:
    * Creates an array of radial distances (r) ranging from slightly outside the event horizon to a larger distance.
5. Orbital and Escape Velocities:
    * Calculates the orbital velocity (v_orb) required for a circular orbit at each radial distance.
    * Calculates the escape velocity (v_esc) required to escape the black hole's gravitational pull from each radial distance.
6. Plotting:
    * Creates a plot showing how the orbital and escape velocities change with radial distance.
    * Includes a horizontal line representing the speed of light for comparison.
Key Concepts:
* Schwarzschild Radius: This is the critical radius where the escape velocity equals the speed of light. Nothing, not even light, can escape from within this radius.
* Orbital Velocity: The velocity required for an object to maintain a stable circular orbit around the black hole.
* Escape Velocity: The minimum velocity an object needs to escape the black hole's gravitational pull.
